<font color='blue'>Looking for specific gender trend from 2012-2018 

```python
#To comment on the demographics of the drug deaths, lets 
#Make a dataframe with only demographic relevant columns.
df1 = pd.DataFrame(df,columns=['Year','Age','Sex','Race']).drop([0])

#There are some unknown and 0 values in this dataframe because of which
#the graphs are not properly plotted.,lets remove those.

df1.drop(df1[(df1['Sex'].values=='Unknown')].index,axis=0,inplace=True)
df1.drop(df1[(df1['Sex'].values==0)].index,axis=0,inplace=True)

#Our data is specificaly from 2012 to 2018 so remove the datapoint 
#which is not in this range and maybe wrong/redundant.

df1.drop(df1[(df1['Year'].values==1970)].index,axis=0,inplace=True)
```

```python
#Countplot gives us the count for each gender in that specific year.
"""
As observed the trend of drug related deaths has been increasing in 
both Male and Female. 
We see that the difference between male and female deaths is large
in all the years. 
"""
print(sns.countplot(x="Year",hue="Sex",data=df1))

#To see the year wise breakdown data between the genders we use 
#groupby

print(df1.groupby('Year')['Sex'].value_counts())

#To see the total male deaths from 2012-2018 we use masking to get
#Male data and then use count to count.

print("\nMale Deaths from 2012-2018:",df1[df1.Sex=='Male']['Sex'].count())

print("\nFemale Deaths from 2012-2018:",df1[df1.Sex=='Female']['Sex'].count())

#To find out how much more male deaths are than female deaths.

M2F_percent = ((df1[df1.Sex=='Male']['Sex'].count())/df1[df1.Sex=='Female']['Sex'].count()).round()

print("\nFor 2012-2018 period,male Deaths are approximately: ",M2F_percent," times greater than the Female deaths")
```

<font color='blue'> Figuring out which race has the maximum deaths

```python
#Cleaning the Race data for analysis.
df1.drop(df1[df1['Race'].values==0].index,axis=0,inplace=True)

race_sex_Vals = df1.groupby('Race')['Sex'].value_counts()
print(race_sex_Vals.sort_values(ascending=False))

##Masking to get race=White and gender=Male
df1[(df1.Race=='White')& (df1.Sex=='Male')]['Race'].count() #getting count

White_M_percent = df1[(~(df1.Race=='White'))& (df1.Sex=='Male')]['Race'].count()/(df1[df1.Sex=='Male']['Sex'].count())
(White_M_percent*100).round()
```

<font color='blue'> Lets look at the age groups and cities now.

```python
#First, lets see the max reported age in the period 2012-2018.
#and the minimum age as well. 
max_Y2A = df1.groupby('Year')['Age'].max()
min_Y2A = df1.groupby('Year')['Age'].min()
#I am using matplot here, and plotting the max-min of the age 
#in the years.
plt.plot(max_Y2A)
plt.plot(min_Y2A)

"""
-> As we can see, the max age seems to be around 85+ for 2013 & 2018.
The max death age decreased from 2013 till 2017. 
-> the min age for a accidental death is more or less around 20 years
as per the graph.

"""
```

```python
#Now lets see the age groups of the dataset reported.
#We will divide the Age column into 6 equal groups of age using pd.cut()
#Also giving labels to the age groups as decimal age doesnt make sense.

df1['Age_Groups'] = pd.cut(df1.Age,6,labels = ['{14-26}yrs','{27-38}yrs','{39-50}yrs','{51-62}yrs','{63-75}yrs','{76-87}yrs'])

print(df1.groupby('Age_Groups').size())

##Using seaborn countplot to get the count of Age_groups
sns.countplot(y="Age_Groups",hue="Sex",data=df1)

"""
As per our observation, the average age group of deaths reported is 
beteen 26-38(40). Overall, the active age group of our data seems to 
be from 26 to 63-64
"""
#df1.groupby('Age_Groups').hist()
```

<font color='blue'>Lets look at something interesting in the drugs involved...

```python
## Making a new DataFrame which has the Year and all the drugs columns from df.
df2 = pd.DataFrame(df,columns=['Year','Heroin','Cocaine','Fentanyl','FentanylAnalogue','Oxycodone','Oxymorphone','Ethanol','Hydrocodone','Benzodiazepine','Methadone','Amphet','Tramad','Morphine_NotHeroin','Hydromorphone','OpiateNOS','AnyOpioid']).drop(0)
```

```python
#There are still few cols which are not int64, so need to convert them.

##Since bool objects cannot be replaced, need to convert to str then int
## Y-A, Y POPS are all in teh dataset, converting it to 1
df2.Fentanyl = df2.Fentanyl.astype('str')
df2['Fentanyl'].replace(['Y-A','Y POPS','Y (PTCH)'],1,inplace=True)
df2.Fentanyl = df2.Fentanyl.astype('int64')

##Chaning and replacing for Morphine_Heroin
df2.Morphine_NotHeroin = df2.Morphine_NotHeroin.astype('str')
df2.Morphine_NotHeroin.replace(['YES','STOLE MEDS','NO RX BUT STRAWS','PCP NEG'],1,inplace=True)
df2.Morphine_NotHeroin.replace('N',0,inplace=True)
df2.Morphine_NotHeroin = df2.Morphine_NotHeroin.astype('int64')

##For AnyOpioid
df2.AnyOpioid = df2.AnyOpioid.astype('str')
df2.AnyOpioid.replace('N',0,inplace=True)
df2.AnyOpioid = df2.AnyOpioid.astype('int64')

df2.dtypes

##As per our steps earlier, we have to again remove 1970 year from the 
## column year as we are primarily foucusing on 2012-2018
df2.drop(df2[(df2['Year'].values==1970)].index,axis=0,inplace=True)
         
```

```python
##Here, we are creating a concatmask to identify which drug was involved.
##Thus, we just convert every column to str and concat it to the next drug col value.

df2['ConcatResult'] = df2.Heroin.astype(str)+df2.Cocaine.astype(str)+df2.Fentanyl.astype(str)+df2.FentanylAnalogue.astype(str)+df2.Oxycodone.astype(str)+df2.Oxymorphone.astype(str)+df2.Ethanol.astype(str)+df2.Hydrocodone.astype(str)+df2.Benzodiazepine.astype(str)+df2.Methadone.astype(str)+df2.Amphet.astype(str)+df2.Tramad.astype(str)+df2.Morphine_NotHeroin.astype(str)+df2.Hydromorphone.astype(str)+df2.OpiateNOS.astype(str)+df2.AnyOpioid.astype(str)

df2['NumberOfDrugs'] = df2.Heroin+df2.Cocaine+df2.Fentanyl+df2.FentanylAnalogue+df2.Oxycodone+df2.Oxymorphone+df2.Ethanol+df2.Hydrocodone+df2.Benzodiazepine+df2.Methadone+df2.Amphet+df2.Tramad+df2.Morphine_NotHeroin+df2.Hydromorphone+df2.OpiateNOS+df2.AnyOpioid

##Here we created and added two new columns to df2.
```

```python
##This line plot shows us number of drugs involved for the relevant years 2012-2018
df2.groupby('Year')['NumberOfDrugs'].value_counts().unstack().plot(kind='line')

##For groupby we need to unstack the columns first and then we use matplot to plot with kind-> line.
##Groupby creates a multiple index so unstack kind of helps us separate them.

"""
As observed, the yellow(1), green(2), red(3) are prominent upwards trends
through 2012 to 2018. Thus, we look deeper into these drugs to find out whcih
ones are they?
"""
```

```python
##Here, I am using the concatresult column to figure out which drugs 
## were used for the red, green and yellow lines above. 
#First we do a value count to understand which concat mask has the highest count.
CR_Vals = df2.ConcatResult.value_counts()
#Then we take the total respective mask's total count.
CR_count = df2.ConcatResult.count()
#Here, we are taking out the percent related to number of drugs present.
percent_2 = df2[(df2.ConcatResult=="1000000000000001")|(df2.ConcatResult=="0010000000000001")]['ConcatResult'].count()/CR_count*100
percent_1 = df2[(df2.ConcatResult=="0100000000000000")|(df2.ConcatResult=="1000000000000000")]['ConcatResult'].count()/CR_count*100
percent_3 = df2[(df2.ConcatResult=="1010000000000001")]['ConcatResult'].count()/CR_count*100
print("The number of deaths involving just Cocaine or Heroin are about:",int((round(percent_1))),"%")
print("The number of deaths involving 2 drugs namely (Heroin & Any Opioid) or (Fentanyl & Any Opioid) are about:",int((round(percent_2))),"%")
print("The number of deaths involving 3 drugs namely (Heroin, Fentanyl & Any Opioid)are about:",int((round(percent_3))),"%")
```

```python
#Classification
import sklearn as sk
import sklearn.tree as tree
from IPython.display import Image  
import pydotplus 
```

```python
df4 = df1.copy()
```

```python
df4 = pd.get_dummies(df4, columns=['Year','Age','Age_Groups','Race','DeathCity'],dummy_na=True)
```

```python
df4.head()
```

```python
dt = tree.DecisionTreeClassifier(max_depth=2) 
```

```python
X = df4.drop('Sex',axis=1)
```

```python
Y = df4.Sex
```

```python
dt.fit(X,Y)
```

```python
# This code will visualize a decision tree dt, trained with the attributes in X and the class labels in Y
dt_feature_names = list(X.columns) #Converting the independent variables to a list 
dt_target_names = [str(s) for s in Y.unique()] #take unique values of Y i.e. DogName
tree.export_graphviz(dt, out_file='tree.dot', 
    feature_names=dt_feature_names, class_names=dt_target_names,
    filled=True)  
graph = pydotplus.graph_from_dot_file('tree.dot') #generating a tree
Image(graph.create_png())
```
