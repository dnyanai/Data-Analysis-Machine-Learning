#### Now lets see what kind of data we have in the Title column

```python
for dataset in combine:
    dataset['Title'] = dataset.Name.str.extract(' ([A-Za-z]+)\.', expand=False)

pd.crosstab(train_df['Title'], train_df['Sex'])
```

#### Now we are establishing a uniformity in the Title column and replacing the discrepencies with the closest possible categorical value.

```python
for dataset in combine:
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess','Capt', 'Col',\
     'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')

    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')
    
train_df[['Title', 'Survived']].groupby(['Title'], as_index=False).mean()
```

#### Here, we are mapping the categorical titles to a numerical equivalent and mapping the Title column values to the corresponding numerical values

```python
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5}
for dataset in combine:
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title'] = dataset['Title'].fillna(0)

train_df.head()
```

```python
train_df = train_df.drop(['Name', 'PassengerId'], axis=1)
test_df = test_df.drop(['Name'], axis=1)
combine = [train_df, test_df]
train_df.shape, test_df.shape
```

```python
train_df.head()
```

#### Converting Sex coulmn to Female : 1 and Male : 0

```python
for dataset in combine:
    dataset['Sex'] = dataset['Sex'].map( {'female': 1, 'male': 0} ).astype(int)

train_df.head()
```

#### Create a guess_age array which will include a guess age based on the Pclass and Sex 

```python
guess_ages = np.zeros((2,3))
guess_ages
```

```python
for dataset in combine:
    for i in range(0, 2):
        for j in range(0, 3): ##iterating from each row in each column and for that Sex and Pclass loc dropping na 
            guess_df = dataset[(dataset['Sex'] == i) & \
                                  (dataset['Pclass'] == j+1)]['Age'].dropna()

            # age_mean = guess_df.mean()
            # age_std = guess_df.std()
            # age_guess = rnd.uniform(age_mean - age_std, age_mean + age_std)

            age_guess = guess_df.median()

            # Convert random age float to nearest .5 age
            guess_ages[i,j] = int( age_guess/0.5 + 0.5 ) * 0.5
            
    for i in range(0, 2):
        for j in range(0, 3):
            dataset.loc[ (dataset.Age.isnull()) & (dataset.Sex == i) & (dataset.Pclass == j+1),\
                    'Age'] = guess_ages[i,j]

    dataset['Age'] = dataset['Age'].astype(int)

train_df.head()
```

```python
train_df['AgeBand'] = pd.cut(train_df['Age'], 5) ##Using pd.cut divide the Age column into 5 groups
train_df[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index=False).mean().sort_values(by='AgeBand', ascending=True)
```

```python
for dataset in combine:    ##Making numerical categories and assigning them to various Ages in those groups
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64, 'Age']
train_df.head()
```

```python
train_df = train_df.drop(['AgeBand'], axis=1) ##Since the age is divided into the age band numerical categories we can drop the column 
combine = [train_df, test_df]
train_df.head()
```

```python
for dataset in combine:
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch'] + 1

train_df[['FamilySize', 'Survived']].groupby(['FamilySize'], as_index=False).mean()\
.sort_values(by='Survived', ascending=False)
```

```python
for dataset in combine: ##We are making a new column as IsAlone so that we can drop the Family size related columns 
    dataset['IsAlone'] = 0
    dataset.loc[dataset['FamilySize'] == 1, 'IsAlone'] = 1

train_df[['IsAlone', 'Survived']].groupby(['IsAlone'], as_index=False).mean()
```

```python
train_df = train_df.drop(['Parch', 'SibSp', 'FamilySize'], axis=1) ## Dropping these columns as IsAlone covers it 
test_df = test_df.drop(['Parch', 'SibSp', 'FamilySize'], axis=1)
combine = [train_df, test_df]

train_df.head()
```

```python
for dataset in combine:
    dataset['Age*Class'] = dataset.Age * dataset.Pclass ##Multiplying Age and Pclass columns to find out Age*Class

train_df.loc[:, ['Age*Class', 'Age', 'Pclass']].head(10) ##Show only columns 
```

#### Working on the Embarked column 

```python
freq_port = train_df.Embarked.dropna().mode()[0] ##Filling the NaN values with the mode of non-null Embraked column values 
freq_port
```

```python
for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].fillna(freq_port) ##Filling the Na values with the freq_port values above 
    
train_df[['Embarked', 'Survived']].groupby(['Embarked'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
for dataset in combine: ##Mapping the numerical categories in the Embarked column respectively 
    dataset['Embarked'] = dataset['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)

train_df.head()
```

```python
test_df['Fare'].fillna(test_df['Fare'].dropna().median(), inplace=True) ##Filling up the NaN values with median of non-zero values in the Fare column
test_df.head()
```

```python
train_df['FareBand'] = pd.qcut(train_df['Fare'], 4) ##Using pd.qcut dividing the Fare column into 4 groups 
##Doing a groupby on FareBand and Survived with FareBand column and taking the mean of it, sorting it by FareBand 
train_df[['FareBand', 'Survived']].groupby(['FareBand'], as_index=False).mean().sort_values(by='FareBand', ascending=True)
```

#### After creating different fare bands, we will give numerical categories to different groups of fares as follows: 

```python
for dataset in combine:
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0 ##Giving label 0 ##Using loc command we are checking for places with Fare value less than 7.91
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[ dataset['Fare'] > 31, 'Fare'] = 3
    dataset['Fare'] = dataset['Fare'].astype(int)

train_df = train_df.drop(['FareBand'], axis=1) ##Since we have divided the Fare column into numerical categories we can drop the  Fare Band column 
combine = [train_df, test_df]
    
train_df.head(10)
```
